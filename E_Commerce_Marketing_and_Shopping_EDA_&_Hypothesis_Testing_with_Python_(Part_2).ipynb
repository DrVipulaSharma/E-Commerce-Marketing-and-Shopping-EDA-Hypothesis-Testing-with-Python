### Handling Outliers Using Capping method

```python
# List of numerical columns with right-skewed data
numerical_cols = ['Income', 'Kidhome', 'Teenhome', 'MntWines', 'MntFruits',
                  'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
                  'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
                  'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

# Define the capping threshold: 95th percentile for right-skewed data
high_quantile = 0.90

# Loop over each feature and apply capping for the upper bound
for feature in numerical_cols:
    # Get the 95th percentile value for the feature (upper bound)
    upper_bound = campaign_data[feature].quantile(high_quantile)

    # Apply capping: replace values above the 90th percentile with the upper bound
    campaign_data[feature] = np.where(campaign_data[feature] > upper_bound, upper_bound, campaign_data[feature])

# Visualize the box plot after capping to confirm changes
plt.figure(figsize=(12, 10))

# Plot the capped features in a grid of boxplots
for i, feature in enumerate(numerical_cols, 1):
    plt.subplot(5, 3, i)  # Adjust grid size based on the number of features (4x4 for 14 columns)
    sns.boxplot(x=campaign_data[feature])
    plt.title(f'Box Plot of {feature} After Capping')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

### Hypothesis Testing

### 1. Is income of customers dependent on their education?

####  Hypothesis:
    
- **Null Hypothesis (Hâ‚€):** Income is independent of the level of education.

- **Alternative Hypothesis (Hâ‚):**  Income is dependent on the level of education.

Test Selection: ANOVA (Analysis of Variance) since we are comparing the means of income across multiple education levels (categorical variable: Education, numerical variable: Income).

```python
from scipy import stats

# Group income by education level
education_groups = [campaign_data['Income'][campaign_data['Education'] == level]
                    for level in campaign_data['Education'].unique()]

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(*education_groups)

# Print F-statistic and p-value
print(f"ANOVA F-statistic: {f_stat}, p-value: {p_value}")

# Set significance level
alpha = 0.05

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis. Income is dependent on education level.")
elif p_value == alpha:
    print("The p-value is equal to the significance level. It's a borderline case.")
else:
    print("Fail to reject the null hypothesis. Income is independent of education level.")

```

### 2. Do higher income people spend more?

#### Hypothesis:

- **Null Hypothesis (Hâ‚€):** Higher-income people do not spend significantly more.
    
- **Alternative Hypothesis (Hâ‚):** Higher-income people spend more across all categories.
    
Test Selection: Pearson's correlation to check the correlation between income (numerical) and total spending (numerical).

```python
from scipy import stats

# Create a total spending column by summing the spending columns
campaign_data['Total_Spending'] = campaign_data[['MntWines', 'MntFruits', 'MntMeatProducts',
                                                 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# Calculate the Pearson correlation between 'Income' and 'Total_Spending'
correlation, p_value = stats.pearsonr(campaign_data['Income'], campaign_data['Total_Spending'])

# Print the results
print(f"Pearson Correlation: {correlation}, p-value: {p_value}")

# Set significance level (alpha)
alpha = 0.05

# Check if the result is statistically significant
if p_value < alpha:
    print(f"The correlation is statistically significant (p-value < {alpha}).")
else:
    print(f"The correlation is NOT statistically significant (p-value >= {alpha}).")
```

- This indicates that there is a strong and significant positive relationship between customers' income levels and their total spending on the specified categories.

### 3. Do couples spend more or less money on wine than people living alone?

#### Hypothesis:

- **Null Hypothesis (Hâ‚€):** There is no difference in wine spending between couples and people living alone.

- **Alternative Hypothesis (Hâ‚):** There is a significant difference in wine spending between couples and people living alone.

Test Selection: An independent t-test since we are comparing the means of two groups (couples vs. alone) on a numerical variable (MntWines).

```python
from scipy import stats

# Create a new column for marital status group: Couples vs. Alone
campaign_data['Living_Status'] = campaign_data['Marital_Status'].replace({
    'Married': 'Couple',
    'Together': 'Couple',
    'Divorced': 'Alone',
    'Single': 'Alone',
    'Widow': 'Alone',
    'Absurd': 'Alone',
    'YOLO': 'Alone'
})

# Filter wine spending for couples and alone
couples_wine_spending = campaign_data['MntWines'][campaign_data['Living_Status'] == 'Couple']
alone_wine_spending = campaign_data['MntWines'][campaign_data['Living_Status'] == 'Alone']

# Perform the t-test (independent t-test for two samples)
t_stat, p_value = stats.ttest_ind(couples_wine_spending, alone_wine_spending)

# Set alpha value
alpha = 0.05

# Print results and interpretation
print(f"T-test statistic: {t_stat}, p-value: {p_value}")

# Check if the p-value is less than alpha to determine statistical significance
if p_value < alpha:
    print("The difference in wine spending between couples and individuals living alone is statistically significant (p < 0.05).")
else:
    print("There is no statistically significant difference in wine spending between couples and individuals living alone (p >= 0.05).")
```

### 4. Are people with lower income more attracted to campaigns?

#### Hypothesis:

- **Null Hypothesis (Hâ‚€):** People with lower income are not more likely to accept campaigns.
    
- **Alternative Hypothesis (Hâ‚):** People with lower income are more likely to accept campaigns.
    
Test Selection: Chi-square test to check the independence between income level (low vs. high) and campaign acceptance (binary: Accepted_Campaign).


```python
from scipy import stats

# Create a column for high and low-income brackets
median_income = campaign_data['Income'].median()
campaign_data['Income_Bracket'] = ['Low' if income <= median_income else 'High' for income in campaign_data['Income']]

# Create a binary column for campaign acceptance (whether any campaign was accepted)
campaign_data['Accepted_Campaign'] = (campaign_data[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3',
                                                     'AcceptedCmp4', 'AcceptedCmp5']].sum(axis=1) > 0).astype(int)

# Create contingency table
contingency_table = pd.crosstab(campaign_data['Income_Bracket'], campaign_data['Accepted_Campaign'])

# Perform Chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Set alpha value
alpha = 0.05

# Print results
print(f"Chi-square statistic: {chi2_stat}, p-value: {p_value}, Degrees of freedom: {dof}")
print(f"Expected frequencies:\n{expected}")

# Check if the p-value is less than alpha to determine statistical significance
if p_value < alpha:
    print("There is a statistically significant association between income brackets and campaign acceptance (p < 0.05).")
else:
    print("There is no statistically significant association between income brackets and campaign acceptance (p >= 0.05).")
```
